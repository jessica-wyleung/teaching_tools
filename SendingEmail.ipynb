```python
import os
import pandas as pd
import smtplib 
from email.mime.multipart import MIMEMultipart 
from email.mime.text import MIMEText 
from email.mime.base import MIMEBase 
from email import encoders 
if __name__ == '__main__':
    xl = pd.ExcelFile(" ")  #Excel file containing student info
    df = xl.parse(sheet_name='') #which sheet to parse
    for filename in os.listdir(""): #directory containing the files you want to attach
        #Extract the file name which is GradeScrope submission ID
        file_ID = filename.split('.')[0]
        current_student = df[df['ID'] == file_ID]
        if len(current_student)>0:
            receiver_email_address = current_student['Email'].astype('str') #Get email address
            SID = str(current_student['SID'].values[0]) #Get SID 
            # Python code to illustrate Sending mail with attachments 
            # from your Uni account  
            fromaddr = " "#"EMAIL address of the sender"
            toaddr = str(receiver_email_address.values[0]) #"EMAIL address of the receiver"

            # instance of MIMEMultipart 
            msg = MIMEMultipart() 

            # storing the senders email address   
            msg['From'] = fromaddr 

            # storing the receivers email address  
            msg['To'] = toaddr 

            # storing the subject  
            msg['Subject'] = "SUBJECT OF EMAIL"

            # string to store the body of the mail 
            body = "Dear student (SID: {}), \n \n TESTING: YOUR MESSAGE HERE. \n \n Regards, \n QBUS Admin Team \n \n".format(SID)

            # attach the body with the msg instance 
            msg.attach(MIMEText(body, 'plain')) 

            # open the file to be sent  
            filename = filename
            dir_filename = str("") #directory to the file you want to attach
            attachment = open(dir_filename, "rb") 

            # instance of MIMEBase and named as p 
            p = MIMEBase('application', 'octet-stream') 

            # To change the payload into encoded form 
            p.set_payload((attachment).read()) 

            # encode into base64 
            encoders.encode_base64(p) 

            p.add_header('Content-Disposition', "attachment; filename= %s" % filename) 

            # attach the instance 'p' to instance 'msg' 
            msg.attach(p) 

            # creates SMTP session 
            s = smtplib.SMTP('smtp.office365.com', 587) 

            # start TLS for security 
            s.starttls() 

            # Authentication 
            s.login(fromaddr, " ")  #YOUR EMAIL PASSWORD GOES HERE

            # Converts the Multipart msg into a string 
            text = msg.as_string() 

            # sending the mail 
            s.sendmail(fromaddr, toaddr, text) 

            # terminating the session 
            s.quit() 
            
            print (SID)
```
